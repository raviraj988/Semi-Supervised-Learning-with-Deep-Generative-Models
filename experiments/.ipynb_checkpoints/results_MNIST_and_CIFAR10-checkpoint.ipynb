```python
import os
import re
import matplotlib.pyplot as plt
import random
import numpy as np
import torch
```

```python
# Paths
script_path = "../src/main.py"
```

```python
SEED = 42
def set_seed(seed):
    """
    Sets the random seed for reproducibility.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(SEED)  # Set the seed globally
```

```python
# Function to run the script and capture validation accuracy
def run_model(command):
    """
    Runs the provided command and captures the validation accuracy from the output.

    Args:
        command (str): Command to run the model training script.

    Returns:
        list: Validation accuracies for each epoch.
    """
    print(f"Running: {command}")
    output = os.popen(command).read()  # Run command and capture output
    accuracies = []

    # Extract validation accuracy from the output
    for line in output.split("\n"):
        match = re.search(r"Validation Accuracy: (\d+\.\d+)%", line)
        if match:
            accuracies.append(float(match.group(1)))
    
    return accuracies
```

```python
# Run the models for MNIST dataset
print("Running on MNIST Dataset...")
m2_command_mnist = f"python {script_path} --dataset MNIST"
optimized_elbo_command_mnist = f"python {script_path} --optimized_elbo --dataset MNIST"

val_acc_m2_mnist = run_model(m2_command_mnist)
val_acc_elbo_mnist = run_model(optimized_elbo_command_mnist)
```

```python
# Run the models for CIFAR-10 dataset
print("\nRunning on CIFAR-10 Dataset...")
m2_command_cifar = f"python {script_path} --dataset CIFAR10"
optimized_elbo_command_cifar = f"python {script_path} --optimized_elbo --dataset CIFAR10"

val_acc_m2_cifar = run_model(m2_command_cifar)
val_acc_elbo_cifar = run_model(optimized_elbo_command_cifar)
```

```python
# Plot validation accuracy vs epoch for MNIST
epochs_mnist = range(1, len(val_acc_m2_mnist) + 1)

plt.figure(figsize=(10, 6))
plt.plot(epochs_mnist, val_acc_m2_mnist, label="M2 Model (MNIST)", marker="o")
plt.plot(epochs_mnist, val_acc_elbo_mnist, label="Optimized ELBO Model (MNIST)", marker="x")
plt.title("Validation Accuracy vs Epoch (MNIST)")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy (%)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot validation accuracy vs epoch for CIFAR-10
epochs_cifar = range(1, len(val_acc_m2_cifar) + 1)

plt.figure(figsize=(10, 6))
plt.plot(epochs_cifar, val_acc_m2_cifar, label="M2 Model (CIFAR-10)", marker="o")
plt.plot(epochs_cifar, val_acc_elbo_cifar, label="Optimized ELBO Model (CIFAR-10)", marker="x")
plt.title("Validation Accuracy vs Epoch (CIFAR-10)")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy (%)")
plt.legend()
plt.grid()
plt.show()
```
